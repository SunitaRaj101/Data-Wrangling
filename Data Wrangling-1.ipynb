```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('penguins_lter.csv')
data.head()
```

```python
data.tail()
```

```python
data.shape
```

```python
#Missing Value Handle
data.isnull()
```

```python
data=data.drop(['Comments'],axis=1)
```

```python
data.shape
```

```python
#Replace missing value with mean
mean=data['Culmen Length (mm)'].mean()
data['Culmen Length (mm)'].replace(np.nan,mean)
```

```python
#replace categorical missing value with frequency
mode=data['Sex'].mode()[0]
print(mode)
data['Sex'].replace(np.nan,mode)
```

```python
#Droping null values
data.dropna(subset=['Date Egg','Culmen Depth (mm)','Flipper Length (mm)','Body Mass (g)','Delta 15 N (o/oo)','Delta 13 C (o/oo)'],axis=0,inplace=True) 
```

```python
#Data Formating
data['Body Mass (g)']=data['Body Mass (g)']/1000
data.rename(columns={'Body Mass (g)':'Body Mass (kg)'},inplace=True)
```

```python
data.shape
```

```python
data
```

```python
data.dtypes
```

```python
#dataframe.astype() to convert datatype
```

```python
#Data Normalization
data['Flipper Length (mm)']=(data['Flipper Length (mm)'].mean())/data['Flipper Length (mm)'].std()
```

```python
data
```

```python
bins=np.linspace(min(data['Body Mass (kg)']),max(data['Body Mass (kg)']),4)
group_name=['low','medium','heavy']
data['Body Mass (kg)-binned']=pd.cut(data['Body Mass (kg)'],bins,labels=group_name,include_lowest=True)
```

```python
import matplotlib.pyplot as plt
plt.hist(data['Body Mass (kg)-binned'])
```

```python
##COnverting Categorical variables into quanititative variables
#ONE HOT ENCODING
dummy_variabl1=pd.get_dummies(data['Sex'],drop_first=False)
dummy_variabl1

#Dummy variable-
```

```python
data.shape
```

```python
#Dummy Encoding
dummy_variable1=pd.get_dummies(data['Sex'],drop_first=True)
dummy_variable1
```

```python
data=pd.concat([data,dummy_variable1],axis=1)
data.drop("Sex",axis=1,inplace=True)
```

```python
#droping missing values
#axis 0 drop entire row, axis 1 drop entire column
print(dataset['fielder'])
dataset.dropna(subset=['fielder'],axis=0,inplace=True)  #dataset=dataset.dropna(subset=["fielder"],axis=0)
print(dataset['fielder'])
```

```python
print(dataset.shape)
```

```python
#Replace missing values with mean
#dataset.replace(missing_value,new_value)
mean=dataset['over'].mean()
print(mean)
dataset['over'].replace(np.nan,mean)
```

```python
#DATA formatting in python or Standardization
print(max(dataset['over']))
dataset['over']=20/dataset['over']
dataset.rename(columns={'over':'avg_over'},inplace=True)
print(dataset['avg_over'])
```

```python
dataset['avg_over'].dtypes
```

```python
#Incorrect data types
#dataframe.astype() to convert datatype
dataset['avg_over']=dataset['avg_over'].astype('int')
dataset['avg_over']
```

```python
#DATA Normalisation- make the data in same range
#normalization on over feature

#simpel featuere scaling:
#dataset['avg_over']=(dataset['avg_over'])/dataset['avg_over'].max()

#Min-Max 
dataset['avg_over']=(dataset['avg_over']-dataset['avg_over'].min())/(dataset['avg_over'].max()-dataset['avg_over'].min())

#Z-Score
# dataset['avg_over']=(dataset['avg_over'].mean())/dataset['avg_over'].std()
```

```python
dataset['avg_over']
```

```python
#Binning :Grouping of values into bins
#converts numeric into categorical variables
#Group a set of numerical values into a set of "bins"


bins=np.linspace(min(dataset['avg_over']),max(dataset['avg_over']),4)
group_name=['low','medium','high']
dataset['over-binned']=pd.cut(dataset['avg_over'],bins,labels=group_name,include_lowest=True)
```

```python
import matplotlib.pyplot as plt
plt.hist(dataset['over-binned'])
```

```python
##COnverting Categorical variables into quanititative variables
#ONE HOT ENCODING

#for Batting team
dummy_variable1=pd.get_dummies(dataset['batting_team'])
dummy_variable1.rename(columns={'Chennai Super Kings':'csk_batting'})
print(dummy_variable1.head)
# dummy_variable2=pd.get_dummies(dataset['bowling_team'])
# dummy_variable2.head('Chennai Super Kings':'csk_bowling')
```

```python
dataset=pd.concat([dataset,dummy_variable1],axis=1)
# dataset=pd.concat([dataset,dummy_variable2],axis=1)
#Drop original column
dataset.drop("batting_team",axis=1,inplace=True)
# dataset.drop("bowling_team",axis=1,inplace=True)
```
